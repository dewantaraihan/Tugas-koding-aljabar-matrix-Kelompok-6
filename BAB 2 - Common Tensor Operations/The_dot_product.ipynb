```python
import numpy as np
import torch
import tensorflow as tf
```

Nomor 1

buat vector x dan y dengan ukuran 1x3 menggunakan numpy

```python
# 1. buat vector x dan y dengan ukuran 1x3
x = np.array([1, 2, 3])
x
```

```python
y = np.array([4, 5, 6])
y
```

buat vector x_pt dan y_pt dengan ukuran 1x3 menggunakan pytorch

```python
x_pt = torch.tensor([1, 2, 3])
x_pt
```

```python
y_pt = torch.tensor([4, 5, 6])
y_pt
```

buat vector x_tf dan y_tf dengan ukuran 1x3 menggunakan tensorflow

```python
x_tf = tf.Variable([25, 2, 5])
x_tf
```

```python
y_tf = tf.Variable([0, 1, 2])
y_tf
```

Nomor 2

hitung perkalian dot product antara vector x dan vector y

```python
np.dot(x, y)
```

hitung perkalian dot product antara vector x_pt dan vector y_pt

```python
torch.dot(x_pt, y_pt)
```

hitung perkalian dot product antara vector x_tf dan vector y_tf

```python
tf.reduce_sum(tf.multiply(x_tf, y_tf))
```

```python
# Mendefinisikan Matrix Y ukuran 2x4
Y = np.array([[42, 4, 7, 99],
              [-99, -3, 17, 22]])
# Menghitung Transpose dari Matrix Y
Y_transpose = Y.T
print(Y_transpose)
```

exercise 2

```python
# Mendefinisikan dua matriks 2x2
matriks_1 = np.array([[25, 10],
                      [-2, 1]])

matriks_2 = np.array([[-1, 7],
                      [10, 8]])
# Menghitung Hadamard Product (Element-wise multiplication)
hadamard_res = matriks_1 * matriks_2
print(hadamard_res)
```

```python
# Mendefinisikan vektor w dan x
w = np.array([-1, 2, -2])
x_val = np.array([5, 10, 0])

# Menghitung Dot Product
dot_wx = np.dot(w, x_val)
print(dot_wx)
```
