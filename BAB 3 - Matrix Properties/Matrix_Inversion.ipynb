1. Buat matrix X ukuran 2x2 menggunakan numpy

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import torch as pt
```

```python
X = np.array([[6, 2], [10, -5]])
X
```

2. Buat invers dari matrix X dan beri nama Xinv menggunakan numpy

```python
Xinv = np.linalg.inv(X)
Xinv
```

3. Kalikan X dengan Xinv menggunakan numpy

```python
np.dot(Xinv, X)
```

4. Selesaikan persamaan berikut menggunakan numpy, tensorflow dan pytorch

Numpy

```python
X = np.array([[4, 2], [-5, -3]])
X
```

```python
Y = np.array([4, -7])
Y
```

```python
solvingNP = np.linalg.solve(X, Y)
solvingNP
```

Tensorflow

```python
X = tf.Variable([[4, 2], [-5, -3]], dtype=tf.float64)
X
```

```python
Y = tf.Variable([[4], [-7]], dtype=tf.float64)
Y
```

```python
solvingTF = tf.linalg.solve(X, Y)
solvingTF
```

Pytorch

```python
X = pt.tensor([[4, 2], [-5, -3]], dtype=pt.float32)
X
```

```python
Y = pt.tensor([[4], [-7]], dtype=pt.float32)
Y
```

```python
solvingPT = pt.linalg.solve(X, Y)
solvingPT
```
